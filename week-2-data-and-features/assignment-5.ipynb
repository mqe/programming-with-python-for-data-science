```python
# TODO:
# Load up the dataset, setting correct header labels
# Use basic pandas commands to look through the dataset...
# get a feel for it before proceeding!
# Find out what value the dataset creators used to
# represent "nan" and ensure it's properly encoded as np.nan

# read in csv file, set index and header
df = pd.read_csv('assets/Datasets/census.data', index_col=0, header=None)

# rename header for columns
df.columns = ['education', 'age', 'capital_gain', 'race', 'capital_loss', 'hours_per_week', 'sex', 'classification']

# reset index of dataset
df.reset_index(drop=True, inplace=True)

# replace 0 with NaN
df = df.replace(0, np.nan)

# check dataset
df
```

```python
# TODO:
# Figure out which features should be continuous + numeric
# Conert these to the appropriate data type as needed,
# that is, float64 or int64

# check data types
df.dtypes
```

```python
# TODO:
# Look through your data and identify any potential categorical
# features. Ensure you properly encode any ordinal types using
# the method discussed in the chapter.

# education
ordered_education = ['7th-8th','9th','10th','11th','12th','HS-grad','Some-college','Bachelors','Masters','Doctorate']
df.education = df.education.astype("category",
  ordered=True,
  categories=ordered_education
).cat.codes

df
```

```python
# TODO:
# Look through your data and identify any potential categorical
# features. Ensure you properly encode any nominal types by
# exploding them out to new, separate, boolean fatures.

# classification
unordered_classification = ['>50K','<=50K']

df.classification = df.classification.astype("category",
  categories=unordered_classification
).cat.codes

# race
unordered_race = ['White','Asian-Pac-Islander', 'Black', 'Amer-Indian-Eskimo', 'Other']

df.race = df.race.astype("category",
  categories=unordered_race
).cat.codes

# sex
unordered_sex = ['Male','Female']

df.sex = df.sex.astype("category",
  categories=unordered_sex
).cat.codes
```

```python
df = pd.get_dummies(df,columns=['classification'])
df = pd.get_dummies(df,columns=['race'])
df = pd.get_dummies(df,columns=['sex'])
```
