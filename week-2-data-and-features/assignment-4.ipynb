```python
import pandas as pd
```

```python
# TODO: Load up the table, and extract the dataset
# out of it. If you're having issues with this, look
# carefully at the sample code provided in the reading
dfs = pd.read_html("http://espn.go.com/nhl/statistics/player/_/stat/points/sort/points/year/2015/seasontype/2", skiprows=0, header=1)
df = dfs[0]
df
```

```python
# TODO: Rename the columns so that they match the
# column definitions provided to you on the website
df.columns = ['rk', 'player', 'team', 'games_played', 'goals', 'assists', 'points', 'plus_minus_rating', 'penalty_minutes', 'points_per_game', 'shots_on_goal', 'shooting_percentage', 'game_winning_goals', 'pp_goals', 'pp_assists', 'sh_goals', 'sh_assists']
```

```python
# TODO: Get rid of any row that has at least 4 NANs in it
df = df.dropna(axis=0, thresh=4)
df
```

```python
# TODO: At this point, look through your dataset by printing
# it. There probably still are some erroneous rows in there.
# What indexing command(s) can you use to select all rows
# EXCEPT those rows?
df = df[df.games_played != "GP"]
df
```

```python
# TODO: Get rid of the 'RK' column
df = df.drop('rk', axis=1)
df
```

```python
# TODO: Ensure there are no holes in your index by resetting
# it. By the way, don't store the original index
df.reset_index(drop=True, inplace=True)
df
```

```python
# TODO: Check the data type of all columns, and ensure those
# that should be numeric are numeric
#df.dtypes
df_test = df.copy()
df_test.games_played = pd.to_numeric(df_test.games_played, errors='coerce')
df_test.goals = pd.to_numeric(df_test.goals, errors='coerce')
df_test.assists = pd.to_numeric(df_test.assists, errors='coerce')
df_test.points = pd.to_numeric(df_test.points, errors='coerce')
df_test.plus_minus_rating = pd.to_numeric(df_test.plus_minus_rating, errors='coerce')
df_test.penalty_minutes = pd.to_numeric(df_test.penalty_minutes, errors='coerce')
df_test.points_per_game = pd.to_numeric(df_test.points_per_game, errors='coerce')
df_test.shots_on_goal = pd.to_numeric(df_test.shots_on_goal, errors='coerce')
df_test.shooting_percentage = pd.to_numeric(df_test.shooting_percentage, errors='coerce')
df_test.game_winning_goals = pd.to_numeric(df_test.game_winning_goals, errors='coerce')
df_test.pp_goals = pd.to_numeric(df_test.pp_goals, errors='coerce')
df_test.pp_assists = pd.to_numeric(df_test.pp_assists, errors='coerce')
df_test.sh_goals = pd.to_numeric(df_test.sh_goals, errors='coerce')
df_test.sh_assists = pd.to_numeric(df_test.sh_assists, errors='coerce')
df_test.dtypes
df_test
```
