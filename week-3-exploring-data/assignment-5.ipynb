```python
import pandas as pd
import matplotlib.pyplot as plt

# import matplotlib
%matplotlib inline
import matplotlib

from pandas.tools.plotting import andrews_curves
from pandas.tools.plotting import parallel_coordinates

# Look pretty...
matplotlib.style.use('ggplot')
```

```python
# read dataset
df = pd.read_csv("assets/Datasets/wheat.data", index_col=0, header=0)
df
```

```python
# create and show andrews curve plot
plt.figure()
andrews_curves(df, 'wheat_type')
plt.show()
```

```python
# dropping 'area' and 'perimeter' features
df = df.drop('area', axis=1)
df = df.drop('perimeter', axis=1)
```

```python
# create and show andrews curve plot on dropped dataset
plt.figure()
andrews_curves(df, 'wheat_type')
plt.show()
```
