# Basic Plots

```python
import pandas as pd

# import matplotlib
%matplotlib inline
import matplotlib
```

```python
# set plot style
matplotlib.style.use('ggplot')
```

## Histogram

```python
# read dataset
student_dataset = pd.read_csv("students.data", index_col=0)
```

```python
# define series and dataframe
my_series = student_dataset.G3
my_dataframe = student_dataset[['G3', 'G2', 'G1']] 
```

```python
# render plots
my_series.plot.hist(alpha=0.5)
my_dataframe.plot.hist(alpha=0.5, normed=True)
```

## 2D Scatter Plots

```python
student_dataset.plot.scatter(x='G1', y='G3')
```

## 3D Scatter Plots

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Final Grade')
ax.set_ylabel('First Grade')
ax.set_zlabel('Daily Alcohol')

# show plot
ax.scatter(student_dataset.G1, student_dataset.G3, student_dataset['Dalc'], c='r', marker='.')
plt.show()
```
