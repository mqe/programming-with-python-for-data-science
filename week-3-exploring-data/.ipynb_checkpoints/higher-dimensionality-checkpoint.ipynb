# Higher Dimensionality

## Parallel Coordinates

```python
from sklearn.datasets import load_iris
from pandas.tools.plotting import parallel_coordinates
import pandas as pd
import matplotlib.pyplot as plt

# import matplotlib
%matplotlib inline
import matplotlib
```

```python
# set plot style
matplotlib.style.use('ggplot')
```

```python
# load sklearn iris dataset into a pandas dataframe
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target_names'] = [data.target_names[i] for i in data.target]
```

```python
# create and show parallel coordinates plot
plt.figure()
parallel_coordinates(df, 'target_names')
plt.show()
```

## Andrew's Curves

```python
from pandas.tools.plotting import andrews_curves
```

```python
# create and show andrews curve plot
plt.figure()
andrews_curves(df, 'target_names')
plt.show()
```

## Imshow

Generates an image based off of the normalized values stored in a matrix, or rectangular array of float64s.

```python
import numpy as np
```

```python
# define dataset as matrix with random floats
df = pd.DataFrame(np.random.randn(1000, 5), columns=['a', 'b', 'c', 'd', 'e'])
df.corr()
```

```python
# create and show imshow plot
plt.imshow(df.corr(), cmap=plt.cm.Blues, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(df.columns))]
plt.xticks(tick_marks, df.columns, rotation='vertical')
plt.yticks(tick_marks, df.columns)
plt.show()
```
